## Data cleaning and preparation

### Subtask:
Handle missing values: Identify and address missing values in the DataFrames, considering appropriate strategies like imputation or removal based on the extent and nature of missingness.


**Reasoning**:
Calculate and display the percentage of missing values for each column in `df_data` and `df_country` to identify columns with high missingness.



```python
missing_data_percentage_data = df_data.isnull().sum() / len(df_data) * 100
print("Missing values percentage in df_data:")
print(missing_data_percentage_data)

missing_data_percentage_country = df_country.isnull().sum() / len(df_country) * 100
print("\nMissing values percentage in df_country:")
print(missing_data_percentage_country)
```

**Reasoning**:
Drop columns from `df_data` and `df_country` with more than 50% missing values, then apply imputation strategies for the remaining columns in both dataframes: forward fill for time-series data in `df_data` and mode imputation for categorical columns in `df_country`.



```python
columns_to_drop_data = missing_data_percentage_data[missing_data_percentage_data > 50].index
df_data = df_data.drop(columns=columns_to_drop_data)
print("Columns dropped from df_data:", columns_to_drop_data)

columns_to_drop_country = missing_data_percentage_country[missing_data_percentage_country > 50].index
df_country = df_country.drop(columns=columns_to_drop_country)
print("Columns dropped from df_country:", columns_to_drop_country)

# Impute missing values in df_data using forward fill for time-series columns
year_columns_data = df_data.columns[4:] # Assuming the first 4 columns are identifiers
df_data[year_columns_data] = df_data[year_columns_data].fillna(method='ffill', axis=1)

# Impute missing values in df_country using mode for categorical columns
for col in df_country.columns:
    if df_country[col].dtype == 'object':
        mode_value = df_country[col].mode()[0]
        df_country[col] = df_country[col].fillna(mode_value)

# Drop rows in df_country where 'Country Code' is missing (although based on the missing percentage, it's 0, this is a good practice)
df_country.dropna(subset=['Country Code'], inplace=True)

print("\nMissing values percentage in df_data after handling:")
print(df_data.isnull().sum() / len(df_data) * 100)

print("\nMissing values percentage in df_country after handling:")
print(df_country.isnull().sum() / len(df_country) * 100)
```

## Data cleaning and preparation

### Subtask:
Reshape the data: The data is currently in a wide format with years as columns. It will be easier to work with if we melt it into a long format with separate columns for 'Year' and 'Value'.


**Reasoning**:
Reshape the wide format `df_data` DataFrame into a long format using melt, focusing on the year columns from '1960' onwards, and convert the 'Year' column to integer type for easier handling.



```python
year_columns = [col for col in df_data.columns if col.isdigit() and int(col) >= 1960]
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']

df_data_long = df_data.melt(id_vars=id_vars,
                            value_vars=year_columns,
                            var_name='Year',
                            value_name='Value')

df_data_long['Year'] = df_data_long['Year'].astype(int)

display(df_data_long.head())
df_data_long.info()
```

## Data cleaning and preparation

Merge data: Combine the data from different DataFrames (e.g., country information and ESG data) into a single DataFrame based on common columns like country codes.

```python
df_merged = pd.merge(df_data_long, df_country, on='Country Code', how='left')
display(df_merged.head())
df_merged.info()
```

## Calculate esg momentum

### Subtask:
Calculate yearly changes


**Reasoning**:
Sort the DataFrame by 'Country Code', 'Indicator Code', and 'Year' to prepare for calculating the yearly change.



```python
df_merged_sorted = df_merged.sort_values(by=['Country Code', 'Indicator Code', 'Year'])
display(df_merged_sorted.head())
```

**Reasoning**:
Calculate the yearly change in the 'Value' column for each country and indicator group and store it in a new column.



```python
df_merged_sorted['Yearly_Change'] = df_merged_sorted.groupby(['Country Code', 'Indicator Code'])['Value'].diff()
display(df_merged_sorted.head())
```

## Calculate esg momentum

### Subtask:
Aggregate momentum: Combine the momentum scores for individual indicators to create an overall ESG momentum score for each country and year.


**Reasoning**:
Calculate the average yearly change for each country and year to get the overall ESG momentum.



```python
df_momentum = df_merged_sorted.groupby(['Country Code', 'Year'])['Yearly_Change'].mean().reset_index()
df_momentum = df_momentum.rename(columns={'Yearly_Change': 'Overall_ESG_Momentum'})
display(df_momentum.head())
```

## Identify high-potential countries

### Subtask:
Rank countries: Based on the calculated ESG momentum, rank countries to identify those with the highest positive momentum.


**Reasoning**:
Calculate the average overall ESG momentum for each country across all years and display the top 10 countries.



```python
average_momentum_per_country = df_momentum.groupby('Country Code')['Overall_ESG_Momentum'].mean().reset_index()
ranked_countries = average_momentum_per_country.sort_values(by='Overall_ESG_Momentum', ascending=False)
print("Top 10 countries with the highest average ESG momentum:")
display(ranked_countries.head(10))
```

## Data visualization

### Subtask:
Plot time series: Generate plots to show the trend of ESG momentum over time for selected high-potential countries.

**Reasoning**:
Filter the momentum data to include only the top countries and then generate line plots for each of these countries showing their ESG momentum over time, including labels, title, and a legend.

```python
top_countries = df_filtered_ranked.head(10)['Country Code'].tolist()

df_top_countries_momentum = df_momentum[df_momentum['Country Code'].isin(top_countries)]

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))

for country_code in top_countries:
    country_momentum_data = df_top_countries_momentum[df_top_countries_momentum['Country Code'] == country_code]
    plt.plot(country_momentum_data['Year'], country_momentum_data['Overall_ESG_Momentum'], marker='o', linestyle='-', label=country_code)

plt.xlabel('Year')
plt.ylabel('Overall ESG Momentum')
plt.title('ESG Momentum Trend Over Time for Top High-Potential Countries')
plt.legend(title='Country Code')
plt.grid(True)
plt.show()
```
